```python
from huggingface_hub import hf_hub_download
import torch
from transformers import PatchTSTConfig, PatchTSTForPrediction
import pandas as pd
import numpy as np
from datetime import timedelta
```

```python
def parse_result(prediction):
    return [int(x + 0.5) if x >= 0 else 0 for x in prediction]
```

```python
model = PatchTSTForPrediction.from_pretrained("../spark/model")

num_predictions = 10
context_length = 8


df = pd.read_csv("./data/time_series.csv", parse_dates=["date"]).iloc[-context_length:]
last_date = df['date'].iloc[-1]
df = df.drop(columns=["date"])



for i in range(num_predictions):
    
    display(df)
    
    past_values = torch.tensor(df.values).view(1, 8, 31).float()    
    outputs = model(past_values=past_values)
    
    predictions_tensor = outputs.prediction_outputs[0]
    predictions = predictions_tensor.detach().numpy()
    
    for i, prediction in enumerate(predictions):
        date = last_date + timedelta(days=i + 1)
        prediction_array = parse_result(prediction)
        print(date, prediction_array)
        
        # Aggiungi una nuova riga alla fine del dataframe
        df.loc[len(df)] = prediction_array
        
        # Rimuovi la riga più vecchia (la prima)
        df = df.iloc[1:].reset_index(drop=True)
    
    # Aggiorna last_date per la prossima iterazione di previsione
    last_date += timedelta(days=len(predictions))

    # Copia il dataframe per evitare modifiche non intenzionali
    df = df.copy()
    
```
